```python
import tensorflow as tf
import tensorflow_probability as tfp
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
import tensorflow.compat.v2 as tf
import tensorflow_datasets as tfds
import time

from RBM import RBM
from utils import flatten_image, flatten_image_label
```

Import MNIST dataset

```python
mnist_builder = tfds.builder("mnist")
mnist_builder.download_and_prepare()

mnist_train, mnist_test = mnist_builder.as_dataset(split=['train[:60000]', 'test'])
```

Train dataset on training data

```python
rbm = RBM(200)
rbm.build([784])
num_epochs=100

start_time = time.time()
rbm.train_rbm(mnist_train, flatten_image, num_epochs, 1e-3, 32)
print('++++++++ Total training time: %.2f seconds ++++++++' % (time.time() - start_time))
```

Make figures of pictures from the test set, and their model reproduction by averaging through 100 generated possibilities:

```python
fig1 = plt.figure(figsize=(20, 20))
fig2 = plt.figure(figsize=(20, 20))

for i, (image, label) in enumerate(mnist_test.map(flatten_image_label).batch(1).take(100)):
    ax1 = fig1.add_subplot(10, 10, i + 1)
    ax2 = fig2.add_subplot(10, 10, i + 1)
    #generate average of 100 evaluations
    show_matrix = np.zeros((28,28))
    for j in range(100):
        show_matrix += rbm.forward_gibbs(image).numpy().reshape(28,28).astype(np.float32)
    show_matrix /= 100
    ax1.imshow(show_matrix, cmap=plt.get_cmap("gray"))
    #ax1.imshow(rbm.forward_gibbs(image).numpy().reshape(28,28).astype(np.float32), cmap=plt.get_cmap("gray"))
    ax2.imshow(image.numpy().reshape(28,28).astype(np.float32), cmap=plt.get_cmap("gray"))
    ax1.axis('off')
    ax2.axis('off')
    i+=1
fig1.tight_layout()
fig1.savefig('Solutions.pdf', format = 'pdf')
fig2.tight_layout()
fig2.savefig('Test_numbers.pdf', format = 'pdf')
```

Plot the value of the 784 (28*28) weights connecting to the first 100 hidden nodes:

```python
plt.figure(figsize=(20, 20))
norm_fact = 0
for i, comp in enumerate(rbm.w_gen.numpy()):
    if max(abs(comp)) > norm_fact:
        norm_fact = max(abs(comp))

for i, comp in enumerate(rbm.w_gen.numpy()):
    if i>=100:
        break
    plt.subplot(10, 10, i + 1)
    plt.imshow(comp.reshape((28, 28)), cmap=plt.cm.RdBu,
               interpolation='nearest', vmin=-norm_fact, vmax=norm_fact)
    plt.axis('off')
    i+=1
plt.tight_layout()
plt.savefig('Filters.pdf', format = 'pdf')
```

Prepare for making a TSNE embedding of the model, to visualize how well the unsupervised learning did:

```python
X_train_numpy = tfds.as_numpy(mnist_train)

N = 10000
X_tsne_folded = []
X_tsne = []
labels = []
for x in mnist_train.batch(1).take(N):
    X_tsne_folded.append(tf.squeeze(tf.divide(tf.dtypes.cast(x["image"], tf.float32), 256.0)))
    flat_image, label = flatten_image_label(x)
    X_tsne.append(flat_image)
    labels.append(label)
```

```python
from sklearn.manifold import TSNE
X_valid_compressed = tf.squeeze(rbm.forward(X_tsne))
tsne = TSNE()
X_valid_2D = tsne.fit_transform(X_valid_compressed)
X_valid_2D = (X_valid_2D - X_valid_2D.min()) / (X_valid_2D.max() - X_valid_2D.min())
```

```python
# adapted from https://scikit-learn.org/stable/auto_examples/manifold/plot_lle_digits.html
import matplotlib as mpl
plt.figure(figsize=(10, 8))
cmap = plt.cm.tab10
plt.scatter(X_valid_2D[:, 0], X_valid_2D[:, 1], c=labels, s=10, cmap=cmap)
image_positions = np.array([[1., 1.]])
for index, position in enumerate(X_valid_2D):
    dist = np.sum((position - image_positions) ** 2, axis=1)
    if np.min(dist) > 0.02: # if far enough from other images
        image_positions = np.r_[image_positions, [position]]
        imagebox = mpl.offsetbox.AnnotationBbox(
            mpl.offsetbox.OffsetImage(X_tsne_folded[index], cmap="binary"),
            position, bboxprops={"edgecolor": cmap(labels[index]), "lw": 2})
        plt.gca().add_artist(imagebox)
plt.axis("off")
plt.tight_layout()
plt.savefig('TSNE_embedding.pdf', format = 'pdf')
plt.show()
```
